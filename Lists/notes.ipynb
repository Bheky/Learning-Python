# Lists

- list is a collection of items in a particular order. it allows us to store sets of information in one place. 
- [] indicates a list with individual elements in the list sepersted by commas. 

```python
bicycles = ['trek', 'cannondale', 'redline', 'specialized']
print(bicycles)
```

## Accessing Elements in a List

- elements in a list can be accessed individually 'indexing'. write the name of the list followed by the index of the item enclosed in square brackets.
- Note: when reading access list items, we use the word sub before the number e.g., students[0] would be spoken as "students sub zero".

```python
bicycles = ['trek', 'cannondale', 'redline', 'specialized']
print(bicycles[2])

print(bicycles[1].title())
```

```python
students = ["Mark", "Amber", "Todd", "Anita", "Sandy"]
print(students)

print(students[0])
```

## Looping Through a List

- to access each item in a list, we use a for loop with this syntax
- Note: always remember to indent the code that's to be executed within the loop.

```python
# Define a list of numbers
scores = [88, 92, 78, 90, 84]
for score in scores:
    print(score)
print("Done")
```

## Checking Whether a List Contains An Item

- we use 'in' in an if statement or a variable assignment.

```python
students = ['Mark', 'Amber', 'Todd', 'Anita', 'Sandy']

# Checking if Anita is in the list
has_anita = 'Anita' in students
print(has_anita)

# Checking if Bob is in the list
has_bob = 'Bob' in students
print(has_bob)
```

## Getting The Length Of A List

- we use the function len() to determine how many items are in a list.

```python
students = ['Mark', 'Amber', 'Todd', 'Anita', 'Sandy']
print(len(students))
```

## Adding An Item To The End Of A List

- we use the .append() method to add a new item to the end of a list. 
- we can either use a variable name or a literal value inside the qoutation marks.

```python
# Create a list of strings (names)
students = ['Mark', 'Amber', 'Todd', 'Anita', 'Sandy']

# Add the name Goober to the list
students.append('Gobber')

new_student = 'Amanda'
#Add whatever name is in new_student to the list.
students.append(new_student)

# Print the entire list
print(students)
```

- sometimes we can run a simple test to see whether a name already exists in a list if not then we append it. for example:

```python
student_name = 'Amanda'

# Add student_name but only if not already in the list.
if student_name in students:
    print(student_name + " already in the list")
else:
    students.append(student_name)
    print(student_name + " added to the list.")
```

## Inserting An Item Into A List

- insert() method allows us to add an item to the list in any position. the syntax is: listname.insert(position, item).

```python
# Create a list of strings (names)
students = ['Mark', 'Amber', 'Todd', 'Anita', 'Sandy']

student_name = 'Lupe'
# Add student name to front of the list.
students.insert(0,student_name)

# Show me the new list.
print(students)
```

## Changing An Item In A List

- we can change an item in a list using the "=" assignment operator like we do with variables.
- the syntax is: listname[index]=newvalue

```python
# Create a list of strings (names).
students = ['Mark', 'Amber', 'Todd', 'Anita', 'Sandy']
students[3] = "Hobart"
print(students)
```

## Combining Lists

- use the extend() function to combine two lists into a single list

```python
# Create two lists of Names
list1 = ['Zara', 'Lupe', 'Hong', 'Alberto', 'Jake']
list2 = ['Huey', 'Dewey', 'Louie', 'Nader', 'Bubba']

# Add list2 names to list1.
list1.extend(list2)

# Print list 1.
print(list1)
```

## Removing List Items

- use remove() method to remove any value from the list.
- Note: if the item is in the list multiple times, only the first occurence is removed

```python
# Create a list of strings.
letters = ['A', 'B', 'C', 'D', 'C', 'E', 'C']

# Remove "C" from the list.
letters.remove("C")

# Show me the new list.
print(letters)
```

- if we need to remove all of an item, we can use a while loop to repeat the .remove as long as the item still remains in the list. 
- if we want to remove an item based on its position in the list, use pop() with an index number.
- Note: if we want to remove the last item from the list, we use pop() without an index number. 

```python
# Create a list of strings.
letters = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

# Remove the first item.
letters.pop(0)
# Remove the last item.
letters.pop()

# Show me the new list.
print(letters)
```

- when we pop() an item off the list, we can store a copy of that value in some variable. 

```python
# Create a list of strings.
letters = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

# Make a copy of the first item then remove it from the list.
first_removed = letters.pop(0)
# Make a copy of last list item then remove it from the list.
last_removed = letters.pop()

# Show the new item list.
print(letters)
# Show what had been removed.
print(first_removed + " and " + last_removed + " were removed from the list.")
```

- also a del command can be used to delete any item from a list based on its index number (position). 

```python
# Create a list of strings.
letters = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

# Remove item sub 2.
del letters[2]

print(letters)
```

- we can also use del to delete an entire list.

## Clearing Out A List

- if we want to delete the contents of a list but not the list itself, we use .clear(). 

```python
# Create a list of strings.
letters = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

# Clear the list of all entries.
letters.clear()

# Show me the new list.
print(letters)
```

## Counting How Many Times An Item Appears In A List

- we can use the count() method to count the number of times an item appears in a list. 

```python
# Create a list of strings.
grades = ['C', 'B', 'A', 'D', 'C', 'B', 'C']

# Count the B's
b_grades = grades.count('B')

# Use a variable for value to count.
look_for = 'C'
c_grades = grades.count(look_for)

print("Thre are " + str(b_grades) + " B  grades in the list.")
print("Thre are " + str(c_grades) + " " + look_for + " grades in the list.")

# Count Fs too.
print("There are " + str(grades.count("F")) + " F grades in the list.")
```

- Note: we use the str() function to convert numbers to strings.
