# Functions


- blocks of code designed to do one specific job. 

```python
def greet_user():
    """"Display a simple greeting."""
    print("Hello!")

greet_user()
```

- in the above example of function definition, "def" keyword tells Python that we're defining a function which is titled "greet_user"
- the text in the tripple qoutation marks is a docstring which describes what the function does. 
- the function call "greet_user()" tells Python to execute the code in the function in this case the body "print("Hello!)". 

### Passing Information to a Function

- we can pass information to a function by adding something to the parentheses of the function's definition. 

```python
def greet_user(username):
    """Display a simple greeting."""
    print(f"Hello, {username.title()}!")

greet_user('jesse')
```

### Arguments and Parameters

- the variable "username" in the definition of "greet_user() in what we call a parameter.
    - parameter = is a piece of information the function needs to do its job.
- and the value 'jesse' in "greet_user('jesse')" is an example of an argument. 
    - argument = piece of information that is passed from a function call to function.

### Passing Arguments

- there is a couple ways of passing arguments to functions:

1. Positional Arguments
    - when we call a function, Python matches each argument in the function call with a parameter in the function definition. for example: <below>
    - Note: we can call a function multiple times using the same function call. 

2. Keyword Arguments
    - a keyword argument is a name-value pair that we can pass to a function. 

3. Default Values
    - a default value is used in an instance where an argument value is not provided.
    - Note: the order of the parameters in the function definition have to be changed. Python interprets this a positional argument which is why the first parameter needs to be "pet_name" since it will be matching parameters and arguments in the function call. 
    - in an instance we need to describe any other animal other than a dog, we explicitly specifiy the "animal_type" in the function call. Python will ignore the parameter's default value. 


```python
# Positional Argument
def describe_pet(animal_type, pet_name):
    """Display information about a pet."""
    print(f"\nI have a {animal_type}.")
    print(f"My {animal_type}'s name is {pet_name.title()}.")

describe_pet('hamster', 'harry')


# Keyword Arguments
def describe_pet(animal_type, pet_name):
    """Display information about a pet."""
    print(f"\nI have a {animal_type}.")
    print(f"My {animal_type}'s name is {pet_name.title()}.")

describe_pet(animal_type='hamster', pet_name='harry')


# Default Values
def describe_pet(pet_name, animal_type='dog'):
    """Display information about a pet."""
    print(f"\nI have a {animal_type}.")
    print(f"My {animal_type}'s name is {pet_name.title()}.")

describe_pet(pet_name='willie')
```
